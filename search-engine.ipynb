# Text tokenization, normalization and lemmatization

The next cell contains the methods for text tokenization and normalization. 

```python
# Creation of the list of pattenrs and replacements
replacement_patterns = [ (r'won\'t', 'will not'), (r'can\'t', 'cannot'),
                        (r'i\'m', 'i am'), (r'ain\'t', 'is not'), 
                        (r'(\w+)\'ll', '\g<1> will'), (r'(\w+)n\'t', '\g<1> not'), 
                        (r'(\w+)\'ve', '\g<1> have'), (r'(\w+)\'s', '\g<1> is'),
                        (r'(\w+)\'re', '\g<1> are'), (r'(\w+)\'d', '\g<1> would') ]


class RegexpReplacer(object):
    def __init__(self, patterns=replacement_patterns):
        self.patterns = [(re.compile(regex), repl) for (regex, repl) in patterns]
    
    def replace(self, text):
        s = text        
        for (pattern, repl) in self.patterns:
            (s, count) = re.subn(pattern, repl, s) # same as sub but return a tuple (new_string, number_of_subs_made)
        return s

# Returns a list of lemmatized words using wordNetLemmatizer
def lemmatization(words):
    lemwn = WordNetLemmatizer()
    return [lemwn.lemmatize(x) for x in words]
    
def normalization(text):
    # Replacement
    import re
    newtext = re.sub(r'(“)', '"', text)
    newtext = re.sub(r'(”)', '"', newtext)
    newtext = re.sub('(’)', "'", newtext)

    
    # Substitution of contractors
    replacer = RegexpReplacer()
    newtext = replacer.replace(newtext)

    # Sentences
    from nltk.tokenize import sent_tokenize
    textbyword = sent_tokenize(newtext)


    # Words
    from nltk.tokenize import RegexpTokenizer
    from nltk import word_tokenize
    import itertools
    single_element_list = []

    for sentence in textbyword:
        #arr = RegexpTokenizer("[\w']+|[\.]|[\,]|[\!]|[can't]").tokenize(sentence)
        tknz = TweetTokenizer(strip_handles = True, reduce_len = True)
        arr = tknz.tokenize(sentence)
        for word in arr:
            single_element_list.append(word)


    # No punctuartion and low cap
    x = re.compile('[%s]' % re.escape(string.punctuation))

    #print('[%s]' % re.escape(string.punctuation), '\n', string.punctuation)

    no_punctuation = []
    for token in single_element_list:
            new_token = re.sub('[\.]|[\"]|[\,]|[\!]|[\:]|[\']|[\‘]', '', token)
            new_token = new_token.lower()
            if not new_token == '':
                no_punctuation.append(new_token)

    # Stop words
    from nltk.corpus import stopwords
    stops = set(stopwords.words('english')) #are all lower case

    doc_norm = [token for token in no_punctuation if token not in stops]
    
    doc_norm = lemmatization(doc_norm)
    #print(doc_norm)
    return doc_norm
```

# Text vectorization

The next cell contains the methods for text vectorization (TF-IDF) and search in the documents.

```python
# Gets a list of tokenized texts. Returns a list of TF-IDF vectors and its list of tokenized
# unique words. Both in the same order.
def vectorization(texts):
    # List of unique words
    TFIDF = []
    words = set()
    for x in texts:
        l = set(x)
        words = words.union(l)
    words = list(words)
    for x in texts:
        v = []
        for y in words:
            isin = 0
            for z in texts:
                if y in z:
                    isin += 1
            v.append(x.count(y)/len(x)*np.log(len(texts)/isin))
        TFIDF.append(v)
    return TFIDF, words

# Gets a tokenized text, a list of TF-IDF vectors and a list of tokenized words. Returns the resultant
# TF_IDF query vector. 
def qvectorizer(query, vectors, words):
    Q1 = []
    for i in words:
        Q1.append(0)
    for val in query:
        for x in vectors:
            for y, z in enumerate(x):
                if val == words[y]:
                    Q1[y] += z
    return Q1
```

The TF-IDF technique obtains an index of each word of the text and its relationship with the text itsel. It has a bidimentional metric for the definition of the place each word fits in the occurrences vs value in the entire text relationship. Let's vectorize all documents.

# Search results using cosine similarity measure

The following cell has the method for the ponderation and sorting of the search results using cosine similarity. For this we need to measure the cosine distance between each chapter vector and the query vector. This gives us a numeric aproximation in similarity for each chapter with the given query.

```python
def chaptersearch(vectors, query):
    results = pd.DataFrame(columns = ['Chapter', 'Index'])
    for i, vector in enumerate(vectors):
        cos = np.dot(query, vector) / np.linalg.norm(query) * np.linalg.norm(vector)
        search = {'Chapter': 'Chapter ' + str(i+1), 'Index': cos}
        results = results.append(search, ignore_index = True)
    return results.sort_values(by=['Index'], ascending = False)
```

For this example I will use the following query: "Siddhartha was a spiritually gifted man who lived his life at the shore of a river."

The reader can input a query in the next cell and get their own results. 

```python
query = list(set(normalization(input())))
query_vector = qvectorizer(query, vectors, words)
results = chaptersearch(vectors, query_vector)
results
```

As we can se abobe, now we have a list of each chapter sorted from the most to the least relevant chapter, just as google search. in the next cell we can see the query vector. Each element represents one of the 3142 words the engine considers of inmportance from the book. The elements with a value different from zero are from words that the engine considers relevant and are contained both in the query and in the book as well. The engine asigns to teh query vector the value of each element with the criteria of summing all the values of teh corresponding element of each chapter.

The following cell contains the method for the search in the tweet list. I made some adjustments, for example, I added the entire boddy of the tweet to the results, also I'm showing only the 10 most relevant tweets.

```python
def tweetsearch(vectors, query, tweets):
    results = pd.DataFrame(columns = ['tweet', 'Index'])
    for i, vector in enumerate(vectors):
        cos = np.dot(query, vector) / np.linalg.norm(query) * np.linalg.norm(vector)
        search = {'tweet': tweets[i], 'Index': cos}
        results = results.append(search, ignore_index = True)
    return results.sort_values(by=['Index'], ascending = False).head(10)
```

 I decided to only use the fiirst 100 tweets since it is a very large database and running times ar eternal with the entire data set. For this example I used the following query: "I'm having a problem with my photo here in twitter amf!!!...can't see my face!"

```python
tweets = list(data['text'])
norm_tweets = []
for i, tweet in enumerate(tweets[:200]):
    norm = normalization(str(tweet))
    norm_tweets.append(norm)
vectors, elements = vectorization(norm_tweets)
query = list(set(normalization(input())))
query_vector = qvectorizer(query, vectors, elements)
results = tweetsearch(vectors, query_vector, tweets)
results
```
